Changing date and time from object to datetime

```python
#changing data types
df['Dates'] = pd.to_datetime(df['Date'])
df['xTime'] = pd.to_datetime(df['Time'])
```

```python
df.dtypes
```

```python
df.drop(['Date'], axis=1)
```

```python
df.drop(['Time'], axis=1)
```

Checking Null Values

```python
df.isna()
```

```python
df.rename(columns={'Dates':'Date', 'xTime':'Time'})
```

```python
df['Driver Cancellation Reason'].unique()
```

```python
df['Reason for cancelling by Customer'].unique()
```

```python
df['Incomplete Rides Reason'].unique()
```

```python
df['Incomplete Rides'].unique()
```

```python
#Replacing null values

df['Avg VTAT'] = df['Avg VTAT'].fillna(0)
df['Avg CTAT'] = df['Avg CTAT'].fillna(0)
df['Incomplete Rides'] = df['Incomplete Rides'].fillna(0)
df['Incomplete Rides Reason'] = df['Incomplete Rides Reason'].fillna("Na")
df['Payment Method'] = df['Incomplete Rides Reason'].fillna("No Payment Made")
df['Incomplete Rides'] = df['Incomplete Rides'].astype(bool)
df['Booking Value'] = df['Booking Value'].fillna(0)
df['Ride Distance'] = df['Ride Distance'].fillna(0)
```

```python
df.to_csv("data_cleaned.csv")
```
